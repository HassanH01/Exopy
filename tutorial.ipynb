Here is a quick tutorial to the exoplanet synthetic data generating python package `Exopy`. You can generate lightcurves of an exoplanet with a host star using two different functions.

```python
# importing necessary modules

import numpy as np
import matplotlib.pyplot as plt
from exopy import lightcurve
```

`with_depth()` function takes single argument, the depth of the transit and produces a tuple containing time array and flux array for plotting lightcurve. Keep in mind that the transit depth $0 < \delta < 1$

```python
data = lightcurve.with_depth(0.2)
plt.scatter(data[0], data[1], marker='.')
plt.title('Lightcurve using transit depth')
plt.xlabel('Phase')
plt.ylabel('Normalized Flux')
plt.show()
```

`with_radius()` function takes two arguments, the radius of the exoplanet and the radius of the host star. The function returns a tuple containing time array and flux for plotting lightcurve. Keep in mind that $R_{\text{planet}} < R_{\text{star}}$

```python
data = lightcurve.with_radius(0.5, 1)
plt.scatter(data[0], data[1],  marker='.')
plt.title('Lightcurve using Planet and Star Radius')
plt.xlabel('Phase')
plt.ylabel('Normalized Flux')
plt.show()
```

`with_noise()` function takes four arguments, the period of the exoplanet, the time for the transit, transit depth and time of observance$ 

```python
data = lightcurve.with_noise(10, 4, 0.2, 10)
plt.scatter(data[0], data[1], marker='.')
plt.title('Lightcurve with Random Noise and Period')
plt.xlabel('Phase')
plt.ylabel('Normalized Flux')
plt.show()
```
